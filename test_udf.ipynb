```
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```
df = spark.createDataFrame(
    [
        (1, "foo", "aa"),  # create your data here, be consistent in the types.
        (2, "bar", "bb"),
    ],
    ["id", "label", "name"]  # add your column names here
)
display(df)
```

```
def find_function(*value_sets):
  def find_col(*columns):
    for i, col in enumerate(columns):
      if col in value_sets[i]:
        return 'Y'
    return 'N'
  
  return udf(find_col, StringType())
```

```
a = set(["foo"])
b = set(["bb"])

find_f = find_function(a, b)
```

```
display(df.withColumn("found", find_f("label", "name")))
```

```
find_f = find_function(a)
display(df.withColumn("found", find_f("label")))
```
